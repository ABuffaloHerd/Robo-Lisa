## Replies
Use code below for pulling replies from json files

```python
df_replies = pd.read_csv('replies.csv')
df_replies.head()
```

```python
df_replies = df_replies.drop_duplicates(subset=['ID', 'Reply ID'])
df_replies.head()
```

```python
with open('Example guild - Text Channels - lit server[1113266262345273428].json', encoding='utf-8') as file:
    data = json.load(file)
    
messages = data['messages']
```

```python
for message in messages[:2]:
    print(json.dumps(message, indent=4))
```

```python
df = pd.json_normalize(messages)
print(df.columns)
df.head()
```

```python
dfjson = df[['id', 'type','content','author.name','reference.messageId']]
dfjson.head()
```

```python
dfjson = dfjson.dropna(subset=['content'])
dfjson = dfjson[dfjson['content'].str.strip() != '']
dfjson.head()
```

```python
merged_df = pd.merge(dfjson, dfjson, left_on='id', right_on='reference.messageId', suffixes=('_message', '_reply'))
print(f"rows: {len(merged_df)}")
merged_df.head()
```

```python
lisa_replies_df = merged_df[merged_df['author.name_reply'] == "lisa38"]
print(f"rows: {len(lisa_replies_df)}")
lisa_replies_df.head()
```

```python
final_df = lisa_replies_df[['content_message', 'content_reply']]
print(f"rows: {len(final_df)}")
final_df.head()
```

```python
final_df = final_df.rename(columns={'content_message': 'original_message', 'content_reply': 'lisas_reply'})
print(f"rows: {len(final_df)}")
```

```python
final_df.to_csv('reply2.csv', index=False)
```

```python
#now lets seperate the emojis from the rest. 
df = pd.read_csv('reply2.csv')
df.head()
```

```python
discord_emoji_pattern = re.compile(r":[a-zA-Z0-9_]+:")

def extract_custom_emojis(text):
    match = discord_emoji_pattern.findall(text)
    return match[0] if match else ''

def remove_custom_emojis(text):
    return discord_emoji_pattern.sub('', text)

df['reply_text'] = df['lisas_reply'].apply(lambda x: remove_custom_emojis(x))
df['reply_emojis'] = df['lisas_reply'].apply(lambda x: extract_custom_emojis(x))
```

```python
df.to_csv('emoji.csv', index=False)
```
